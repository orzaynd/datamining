```python
df_excel = df_excel[['Last name', 'First name', 'ID number', 'Grade/100.00']]
```

```python
print(df_mysql['nim'].head(5))
print(df_excel['ID number'].head(5))
```

```python
df_excel['ID number'] = df_excel['ID number'].fillna(0).astype(int).astype(str)
df_excel.head()

#tambahkan "0" di depan ID number yang kurang dari 8 digit
df_excel['ID number'] = "0" + df_excel['ID number']
df_excel.head()
```

```python
df_excel = df_excel.copy()
df_mysql = df_mysql.copy()

df_excel['ID number'] = df_excel['ID number'].astype(str)
df_mysql['nim'] = df_mysql['nim'].astype(str)
```

```python
df_integrasi = pd.merge(df_mysql, df_excel, left_on='nim', right_on='ID number', how='inner')
df_integrasi.head()
```

```python
df_integrasi = df_integrasi[['id', 'nim', 'nama', 'Last name', 'First name', 'Grade/100.00']]

df_integrasi.head()
```

```python
df_integrasi.rename(columns={'Grade/100.00': 'Nilai'}, inplace=True)
df_integrasi.head()
```
