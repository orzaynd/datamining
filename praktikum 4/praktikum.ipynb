```python
df_rumahSakitBesar = df[df['total_tempat_tidur'] > 100 ]
df_rumahSakitBesar.head()
#menampilkan data pada kolom total_tempat_tidur yang bernilai lebih dari 100
```

```python
df_filter = df[(df['propinsi'] == 'DKI Jakarta') & (df['total_tempat_tidur'] > 100)]
df_filter.head()    
#menampilkan data pada kolom provinsi yang bernilai DKI Jakarta dan total_tempat_tidur yang bernilai lebih dari 100     
```

```python
#menampilkan data pada kolom provinsi yang bernilai Papua atau total_tenaga_kerja yang bernilai kurang dari 10
df_filter_RumahSakitKecil = df[(df['propinsi'] == 'Papua') | (df['total_tenaga_kerja'] < 10)]
df_filter_RumahSakitKecil.head()
```
